In this exercise, you'll redo the data gathering phase of the UNData Exploration project by using APIs instead of downloading csv files.

You'll make use of the [World Bank Indicators API](https://datahelpdesk.worldbank.org/knowledgebase/articles/889392-about-the-indicators-api-documentation). Note that this API does not require an API key. Before attempting the exercise, it would be a good idea to skim through the Documentation page and to check out the [Basic Call Structure article](https://datahelpdesk.worldbank.org/knowledgebase/articles/898581).

1. Use the API to get all available data for the _GDP per capita, PPP (constant 2017 international $)_ indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

```python
import requests 
import pandas as pd
import numpy as np
```

```python
params = {"format": "json", "per_page":18000}
    
endpoint = "https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD" 

response = requests.get(endpoint, params = params) 
```

```python
response
```

```python
response.json()[1]
```

```python
GDP_Per_Capita = pd.json_normalize(response.json()[1])
GDP_Per_Capita
```

```python
GDP_Per_Capita = (
    GDP_Per_Capita.drop(columns = ["unit", "obs_status", "decimal", "indicator.id", "indicator.value"])
    .rename(columns = {"value": "GDP Per Capita", "date": "Year", "country.value": "Country"})
)
```

```python
GDP_Per_Capita.head(2)
```

2. Now, use the API to get all available data for _Life expectancy at birth, total (years)_. This indicator has code "SP.DYN.LE00.IN". Again, convert the results to a DataFrame.

```python
params = {"format": "json", "per_page":18000}
    
endpoint = "https://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN"

response = requests.get(endpoint, params = params) 
```

```python
response.json()[1]
```

```python
Life_expectancy_at_birth = pd.json_normalize(response.json()[1])
Life_expectancy_at_birth.head(2)
```

```python
Life_expectancy_at_birth = (
    Life_expectancy_at_birth.drop(columns = ["unit", "obs_status", "decimal", "indicator.id", "indicator.value"])
    .rename(columns = {"value": "Life Expectancy(Years)", "date": "Year", "country.value": "Country"})
)
```

```python
Life_expectancy_at_birth.head(2)
```

3. Merge the two results DataFrames together. You may want to rename or drop columns prior to merging.

```python
Life_expectancy_by_GDP = pd.merge(left=GDP_Per_Capita, 
                                  right=Life_expectancy_at_birth)
Life_expectancy_by_GDP.head(2)
```

```python
Life_expectancy_by_GDP.sort_values('Year', ascending=True)
```

You can also get more information about the available countries (region, capital city, income level classification, etc.) by using the [Country API](https://datahelpdesk.worldbank.org/knowledgebase/articles/898590-country-api-queries). Use this API to pull in all available data. Merge this with your other datasets. Use this to now remove the rows that correspond to regions and not countries.

```python
params = {"format": "json", "per_page":296}
    
endpoint = "https://api.worldbank.org/v2/country/" 

response = requests.get(endpoint, params = params) 
```

```python
response 
```

```python
response.json()[1]
```

```python
Country_MOREdata_df = pd.json_normalize(response.json()[1]).rename(columns={"iso2Code": "country.id", "id": "countryiso3code", "name": "Country"})
Country_MOREdata_df#.rename(columns={"iso2Code": "country.id", "id": "countryiso3code", "name": "Country"})
```

```python
LE_by_GDP_MD = pd.merge(left=Life_expectancy_by_GDP, 
                        right=Country_MOREdata_df,
                        how="inner", 
                        on=["countryiso3code", "country.id", "Country"],
                        validate="m:1")
LE_by_GDP_MD
```

```python
#replace empty rows with NaN values
LE_by_GDP_MD["capitalCity"] = LE_by_GDP_MD["capitalCity"].replace('', np.nan)
#Drops rows where column has NaN values
LE_by_GDP_LD = LE_by_GDP_MD.dropna(subset=["capitalCity"])
LE_by_GDP_LD
```
